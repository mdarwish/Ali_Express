```python
import re
import os
import pandas as pd
import time
from time import sleep 

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pyautogui
from bs4 import BeautifulSoup

base_url = 'https://www.aliexpress.com/item/'
# countries dropdown tag: input, id: ae-search-select-1
```

```python
products_list = [
                 '32897621367',
                 '32851991552',
                 '32813349070',
                 '32905498695',
                 '32889687158', # only Canada
                 '32948609346',
                 '2055674172'
                ]
```

```python
products_soups = []
products_data = []
```

```python
# chrome_folder = 'D:\\Documents\\AliExpress\\Products\\Chrome\\'
target_folder = '/media/dodz/Data/Documents/AliExpress/Products/Chrome/'
# lmedialdodzlDatalDocumentslAliExpresslProductslChromel32685273775
delay = 3 # seconds

print("Target folder: {}".format(target_folder))

# if save folder doesn't exist, create
if not os.path.exists(target_folder):
    os.mkdir(target_folder)

# create a new Chrome session
browser = webdriver.Firefox(executable_path='/media/dodz/Data/Linux/web_drivers/geckodriver')
for product in products_list:
    url = base_url + product + '.html'
    browser.get(url)
    time.sleep(delay)
    try:
        browser.find_elements(by='xpath',value="//a[@role='button']")[0].click()
    except:
        None

    time.sleep(delay)
    browser.find_elements(by='xpath',value="//span[@class='product-shipping-info black-link']")[0].click()
    time.sleep(delay)
    shipping_options = browser.find_elements_by_xpath("//div[@role='radiogroup']")
    for service in shipping_options:
        if service.text.find('ePacket') != -1:
            service.click()
    shipping_options = shipping_options[0].text
    sleep(delay)
    browser.find_elements(by='xpath',value="//button[@class='next-btn next-medium next-btn-primary']")[0].click()
    
    scroll = 100
    for i in range(1,100):
        
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight/{});".format(scroll-i))
        sleep(0.1)
        i += 10
    
    product_soup = BeautifulSoup(browser.page_source, "html.parser")
    products_soups.append(product_soup)
    title = product_soup.find('div', {'class':'product-title'})
    price = product_soup.find('span', {'class':'product-price-value'})
    destination = product_soup.find('span', {'class':'product-shipping-info black-link'})
    time.sleep(delay)    
    product_overview = product_soup.find('div', {'class':'product-overview'})
    paragraphs = product_overview.find_all('p') + product_overview.find_all('table')
    data_record = [product, title.text, price.text, shipping_options, destination.text, [paragraph.text for paragraph in paragraphs], product_soup]
    products_data.append(data_record)

    pyautogui.hotkey('ctrl', 's')
    time.sleep(3)

    filename = target_folder + product
    print(filename)
    pyautogui.typewrite(filename)
    pyautogui.hotkey('enter')
    time.sleep(5)

time.sleep(5)
browser.close()
```

```python
from PIL import Image
```

```python
# save_folder = 'D:\\Documents\\AliExpress\\Images'
save_folder = r'/media/dodz/Data/Documents/AliExpress/Images'

# base_dir = 'D:\\Documents\\AliExpress\\Products\\Chrome'
base_dir = r'/media/dodz/Data/Documents/AliExpress/Products'


# if save folder doesn't exist, create
if not os.path.exists(save_folder):
    os.mkdir(save_folder)

for subdir, dirs, files in os.walk(base_dir):
    counter = 1
    # print (subdir, dirs, files)
    for file in files:
        filepath = subdir + os.sep + file
        match = re.search(r'\/(?P<sku>\d*)_files$', str(subdir), re.I | re.U)
        if match:
            product = match.group('sku')
        
        if filepath.endswith(".jpg") or filepath.endswith(".png") or filepath.endswith(".webp"):
            im = Image.open(filepath)
            if not im.mode == 'RGB':
                im = im.convert('RGB')
            if im.size[0] > 200 and im.size[1] > 200:
                new_file = save_folder + os.sep + product + "_" + str(counter) + ".jpg"
                print (new_file)
                im.save(new_file, "JPEG", quality=95)
                counter += 1
                
```
